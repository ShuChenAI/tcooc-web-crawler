```python
from typing import Tuple

import requests

JINA_API_URL = 'https://r.jina.ai/'

with open('jina_api_key.txt', 'r') as f:
    JINA_API_KEY = f.read().strip()

JINA_HEADERS = {
    'Accept': 'application/json',
    'Authorization': f'Bearer {JINA_API_KEY}',
    'X-No-Cache': 'true',
    'X-With-Links-Summary': 'true',
    'X-With-Generated-Alt': 'true'
}


def jina_reader_task(web_url: str) -> Tuple[str, str]:
    payload = dict(url=web_url)
    response = requests.post(JINA_API_URL, headers=JINA_HEADERS, data=payload)

    if response.status_code != 200:
        return '', ''

    data = response.json().get('data', {})

    web_title = data.get('title', 'No Title')
    web_content = data.get('content', 'No Content')

    return web_title, web_content
```

```python
from concords_web_rpa.browser import WebBrowser
from selenium.webdriver.common.by import By
from pathlib import Path

browser = WebBrowser(ads_id='kl4stp4', port=50360)
```

```python
browser.driver.get("https://tcooc-co.gov.taipei/News.aspx?n=71C48CD9E02C4FE2&sms=14CD04DB05B73A14&page=1&PageSize=500")

root = "公司登記事宜"
Path(root).mkdir(exist_ok=True)
```

```python
table_xpath = '//*[@id="table_0"]/tbody'
table = browser.driver.find_element(By.XPATH, table_xpath)
rows = table.find_elements(By.XPATH, 'tr')
```

```python
links = []
for row in rows:
    question = row.find_elements(By.XPATH, 'td')[1]
    url = question.find_element(By.XPATH, 'span/a').get_attribute('href')
    links.append(url)
```

```python
for idx, url in enumerate(links):
    browser.driver.get(url)
    content = browser.get_element('//*[@id="CCMS_Content"]/div').text
    print(content)
    with open(f"{root}/{idx}.txt", 'w', encoding='utf-8') as f:
        f.write(content)
```
